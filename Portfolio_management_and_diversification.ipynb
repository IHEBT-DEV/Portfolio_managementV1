### APIExtractor class
>This class is used to import data using the **requests** library, we can utilize its functionalities for making HTTP requests.

It's crucial to note that Coinbase APIs impose a **limit** of **300 data points per single request**. For instance, when utilizing **hour** granularity, a single request provides data for **only 12.5 days**. To overcome this limitation, we address the point constraint by sending **multiple requests** with **chunked** start and end date ranges.

````python
#get data using requests
def get_data(base_url, endpoint, params=None, headers=None):
  try:
    response = requests.get(f"{base_url}/{endpoint}", headers=headers, params=params)
    response.raise_for_status()  # Raise an exception if the request was not successful

    if response.status_code == 200:
        return response.json()  # Assuming the API response is in JSON format
    else:
        print(f"API request failed with status code: {response.status_code}")
        return None
  except requests.exceptions.RequestException as e:
    print(f"API request error: {e}")
    return None

#Api extractor class
class APIExtractor:
  def __init__(self):
    self.get_data = get_data
````

### DateHelpers class
>This class encapsulates essential functions and algorithms for date manipulation.

````python
#the number of days between two timestamps
def number_of_days(timestamp1, timestamp2):
    delta_days = int(abs(int(timestamp1) - int(timestamp2)) / (60 * 60 * 24))
    return delta_days
#Number of points needed to be imported by start and end dates
def number_of_points(start_date, end_date, granularity):
    day_to_s_coefficient = 86400
    nb_days = number_of_days(start_date, end_date)
    nb_points = int(nb_days * day_to_s_coefficient / granularity)
    return nb_points
#start and end date chunk
def list_start_end_chunks(start_date, end_date, granularity):
    nb_points = number_of_points(start_date, end_date, granularity)
    list_starts_end_dates = []
    if nb_points <= 300:
        return [{'start_date': start_date, 'end_date': end_date}]
    else:
        ms_delay = int(299 * granularity)
        nb_seconds=end_date-start_date
        for i in range(start_date, end_date,ms_delay):
            list_starts_end_dates.append(
                {'start_date':  i ,
                 'end_date': i+ms_delay-granularity if i+ms_delay<end_date else end_date
                 })
        return list_starts_end_dates

class DateHelpers:
    def __int__(self):
        self.list_start_end_chunks = list_start_end_chunks
````

>This class is designed to parse raw candlestick data obtained from the Coinbase API, transforming it into a structured dictionary format for easier handling and analysis.

````python
#this function calls the spot parser function by market and currency
def parse_spot_by_market_currency(data, market, currency):
    if market.upper() == "COINBASE":
        return coinbase_ohlcv_parser(data, currency)
#coinbase ohlcv parser
def coinbase_ohlcv_parser(data, currency):
  parsed_coinbase_spot_data = []
  for mydata in list(data):
    date_to_datetime = datetime.datetime.fromtimestamp(int(mydata[0]))
    parsed_coinbase_spot_data.append({
        'currency': f"{currency}USDT",
        'date': date_to_datetime,
        'low': float(mydata[1]),
        'high': float(mydata[2]),
        'open': float(mydata[3]),
        'close': float(mydata[4]),
        'volume': float(mydata[5])
    })
  return parsed_coinbase_spot_data

class CandlestickParser:
  def __init__(self):
    self.parse_spot_by_market_currency = parse_spot_by_market_currency
````

### SpotServices Class
>This class comprises spot services.

 Presently, it features a function for importing historical spot data from Coinbase through the API and preparing structured parsed data. In the future, we have the capability to save the resulting data into CSVs, different databases, or the AWS S3 bucket.

````python
#get historical spot data from coinbase
def get_spot_coinbase_api(currency='BTC', granularity=3600, start_date=None, end_date=None):
  coinbase_api_url = f"https://api.exchange.coinbase.com/products/{currency}-USDT"  # coinbase base url

  spot_endpoint = "candles"  # candlestick endpoint
  #api extractor instance
  api_extractor_instance = APIExtractor()
  data = []
  if start_date:
    if not end_date :
      end_date = int(time.time())
    list_dates = list_start_end_chunks(start_date=start_date, end_date=end_date, granularity=granularity)
    for my_dates in list(list_dates):
      spot_params = {"granularity":int(granularity),
                      'start': my_dates['start_date'],
                      'end': my_dates['end_date']}
      data_chunk = api_extractor_instance.get_data(coinbase_api_url, spot_endpoint, spot_params)
      data.append(data_chunk)
  #last 300 points
  else:
    spot_params = {"granularity":int(granularity)}
    data = api_extractor_instance.get_data(coinbase_api_url, spot_endpoint, spot_params)
  return data

#this function calls the needed get historical spot dat by market (for now, we have only coinbase)
def get_spot_api(market='coinbase', currency='BTC', granularity=86400, start_date=None, end_date=None):
    if market.upper() == 'COINBASE':
        return get_spot_coinbase_api(currency, granularity, start_date, end_date)

#Prepare data frame with close as values and currencies as columns indexed by date
def close_df_preparation(ohlc_df):
  close_df = ohlc_df.reset_index().pivot(index = 'date', columns = 'currency', values = 'close')
  #Na handling
  close_df=close_df.ffill()
  close_df=close_df.bfill()
  close_df.index = pd.to_datetime(close_df.index)
  return close_df

class SpotServices:
  def __init__(self):
      self.get_spot_api = get_spot_api
      self.close_df_preparation=close_df_preparation
  def historical_spot(self,market,currency,granularity,start_date=None,end_date=None):
    #get all spot data
    spot_data_all = self.get_spot_api(market=market.upper(), currency=currency, granularity=granularity,
                                                    start_date=start_date, end_date=end_date)
    spot_candlestick_parser = CandlestickParser()
    parsed_coinbase_data = []
    #if data gathered from async (multiple spot data lists)
    if type(spot_data_all[0]) is list :
      for spot_data in spot_data_all:
        parsed_coinbase_data = parsed_coinbase_data + spot_candlestick_parser.parse_spot_by_market_currency(spot_data,
                                                                                                        market,
                                                                                                        currency)
    #one spot data list
    else:
      parsed_coinbase_data = parsed_coinbase_data + spot_candlestick_parser.parse_spot_by_market_currency(spot_data_all,
                                                                                                        market,
                                                                                                        currency)
    return {'data': parsed_coinbase_data}
  #Prepare ohlcv data frame
  def ohlcv_df_preparation(self,assets,granularity,start_date):
    currencies = sorted(assets)
    all_dfs=[]
    for currency in currencies :
      data_i = self.historical_spot('coinbase',currency,granularity,start_date)
      df_i = pd.DataFrame(data_i['data'])
      df_i.sort_values(by=['date'], ascending = False, inplace = True)
      df_i.set_index(['currency','date'], inplace=True)
      all_dfs.append(df_i)
    df=pd.concat(all_dfs,axis=0)
    return df
````

##Some portfolio basic concepts



>   **Profitability concept** : The profitability of the portfolio is determined by the linear combination of the returns of the assets it comprises.
\begin{equation}
\widetilde{r}_P=\sum_{i=1}^n \alpha_i \widetilde{r}_i \text { Where } \sum_{i=1}^n \alpha_i=1
\end{equation}

>  **Yield/return concept** : It is the mathematical expectation of returns.
\begin{equation}
\mu=E\left(\widetilde{r}_P\right)=\sum_{i=1}^n \alpha_i \mu_i
\end{equation}

>   **Risk concept** : The risk of the portfolio is measured by the square root of the variance of its returns (standard deviation) ; It must consider the risk of each assets as well as their joint movement, which either amplifies or mitigates the overall risk.
\begin{equation}
\sigma^2=V\left(\widetilde{r}_P\right)=\sum_{i=1}^n \sum_{j=1}^n \alpha_i \alpha_j \rho_{i, j} \sigma_i \sigma_j
\end{equation} Where \begin{equation}
\sigma_{i, j}=\rho_{i, j} \sigma_i \sigma_j: \text {covariance between } A_i \text { and } A_j
\end{equation}

>**Annualization concept**: To annualize returns/risk is to convert returns/risk from a shorter time period (e.g., quarter daily in our case) to an annualized rate. The annualization process is essential for comparing returns/risk across different time frames consistently. The method of annualization on the frequency.
 * **Annualized Return formula:**
\begin{equation}
\text { Annualized Return }=(1+\text { Average granularity Return })^{\text {Number of Periods per Year }}-1
\end{equation}
 * **Annualized volatility/risk formula:**
\begin{equation}
\text { Annualized Volatility }=\text { Volatility of Returns } \times \sqrt{\text { Number of Periods per Year }}
\end{equation}

### PortfolioService class
>This class contains portfolio services.

````python
'''
################################################################################
                            Annualization factor
################################################################################
'''
# this funtion returns the yearly number of periods
def yearly_nb_periods(granularity=21600):
  day_s = 86400
  trading_days=365
  daily_nb_period = day_s/granularity
  yearly_nb_periods = trading_days*daily_nb_period
  return yearly_nb_periods
'''
################################################################################
                    Portfolios return/risk df
################################################################################
'''
#this function returns a dataframe with return and volatility of each asset
def asset_returns_volatility(expected_returns,annualized_volatility):
    assets = pd.concat([expected_returns, annualized_volatility], axis = 1)
    assets.columns = ['Returns', 'Volatility']
    return assets

class PortfolioService:
  def __init__(self,close_df):
    self.asset_returns_volatility=asset_returns_volatility
    self.yearly_nb_periods=yearly_nb_periods
    self.close_df=close_df
  '''
  ##############################################################################
                            Return/risk annualization
  ##############################################################################
  '''
  #expected  return annualization
  def expected_return_annualized(self):
    annualizing_factor = yearly_nb_periods()
    expected_returns = self.close_df.pct_change().bfill().mean().apply(lambda x :  (1 + x)**annualizing_factor - 1)
    return expected_returns
  #volatility annualization
  def volatility_annualized(self):
    annualizing_factor = yearly_nb_periods()
    annualized_volatility = self.close_df.pct_change().std().apply(lambda x: x*np.sqrt(annualizing_factor))
    return annualized_volatility
  '''
  ##############################################################################
                              Assets covariance matrix
  ##############################################################################
  '''
  #covariance matrix
  def covariance_matrix(self):
      cov_matrix = self.close_df.pct_change().ffill().cov()
      return cov_matrix
  '''
  ##############################################################################
                     Portfolios with random weights construction
  ##############################################################################
  '''
  #this function simulates different random portfolios
  def portfolios_simulation(self,expected_returns,cov_matrix,num_port=1000):
      p_return = []
      p_vol = []
      p_weights = []
      annualizing_factor = yearly_nb_periods()
      num_assets = len(self.close_df.columns)
      for p in range(num_port):
          weights = np.random.random(num_assets)
          weights = weights / np.sum(weights)
          p_weights.append(weights)
          returns = np.dot(weights, expected_returns)
          p_return.append(returns)
          var= cov_matrix.mul(weights, axis = 0).mul(weights, axis = 1).sum().sum()
          sd= np.sqrt(var)
          annualized_sd = sd * np.sqrt(annualizing_factor)
          p_vol.append(annualized_sd)
      data = {'Returns': p_return, 'Volatility': p_vol}
      for counter, symbol in enumerate(close_df.columns.tolist()):
          data[f'{symbol}_weight'] = [w[counter] for w in p_weights]
      portfolios = pd.DataFrame(data).sort_values('Returns', ascending = False)
      return portfolios
````

## Efficient Frontier


We refer to the efficient frontier as the boundary or curve encompassing portfolios that achieve the maximum return for a given level of risk or the minimum risk for a given level of return. In essence, it represents the line that offers the optimal trade-off between return and risk in portfolio construction.

###Markowitz Mean-Variance Portfolio Optimization



Markowitz Mean-Variance Portfolio Optimization is a framework for constructing investment portfolios that aims to achieve the optimal balance between risk and return. Developed by Harry Markowitz, this approach considers the expected returns and the covariance of asset returns to create portfolios that offer the highest expected return for a given level of risk or the lowest risk for a given level of return.

**Mathematical concepts** :  
>**Objective Function:**
The objective of portfolio optimization is to maximize the Sharpe ratio. The Sharpe ratio is defined as the ratio of the portfolio's excess return to its volatility (risk).  The Sharpe ratio (SR) is given by:
$$
S R=\frac{R_p-R_f}{\sigma_p}
$$
where:
- $R_p$ is the expected portfolio return.
- $R_f$ is the risk-free rate.
- $\sigma_p$ is the standard deviation (volatility) of the portfolio's return.

>**Constraints:**
There is one equality constraint in the optimization problem, ensuring that the sum of weights in the portfolio equals 1. This constraint is expressed as:
$$
\sum_{i=1}^n w_i=1
$$
where:
- $w_i$ is the weight of asset $i$.

>**Optimization Method:**
The optimization is performed using the Sequential Least Squares Quadratic Programming (SLSQP) method. The minimize function from scipy.optimize is employed for this purpose.

>**Portfolio Generation:**
Portfolios along the efficient frontier are generated by varying the target return levels. For each target return, the optimization problem is solved to find the weights that maximize the Sharpe ratio while satisfying the sum of weights constraint.

Needed library

Scipy.optimize, part of the Scipy library, offers a collection of optimization algorithms for solving unconstrained and constrained mathematical optimization problems.
It includes methods for root finding, linear programming, and global optimization, making it a versatile tool for scientific and data analysis applications.

````python
# Constraint: weights must sum to 1
constraints = ({'type': 'eq', 'fun': lambda w: np.sum(w) - 1})
class EfficientFrontierService:
  def __init__(self,returns,cov_matrix):
      self.returns=returns
      self.cov_matrix=cov_matrix
      self.constraints = constraints
  '''
  ##############################################################################
                            Objective function
  ##############################################################################
  '''
  # Objective function for portfolio optimization (maximize Sharpe ratio)
  def objective_function(self,weights):
      annualizing_factor = yearly_nb_periods()
      portfolio_return = np.sum(self.returns * weights)
      portfolio_volatility = np.sqrt(np.dot(weights.T, np.dot(self.cov_matrix, weights)))*np.sqrt(annualizing_factor)
      obj =  portfolio_return/portfolio_volatility
      return -obj
  '''
  ##############################################################################
               Efficient Portfolios with optimal  weights construction
  ##############################################################################
  '''
  #This function constracts efficient frontier portfolios
  def efficient_frontier_portfolios(self):
      # Initial guess for weights (equal weights)
      initial_weights = np.ones(len(self.returns)) / len(self.returns)
      # Generate portfolios along the efficient frontier
      target_returns = np.linspace(min(self.returns), max(self.returns), num=1000)
      portfolios = []
      for target_return in target_returns:
          constraints = ({'type': 'eq', 'fun': lambda w: np.sum(w) - 1},
                      {'type': 'eq', 'fun': lambda w: np.sum(self.returns * w) - target_return}
                         )
          result = minimize(self.objective_function, initial_weights,
                          method='SLSQP', constraints=constraints)
          portfolios.append(result.x)
      return portfolios
  '''
  ##############################################################################
                           Portfolio return and risk
  ##############################################################################
  '''
  #this function calculates portfolios returns
  def portfolio_returns(self,portfolios):
      portfolio_returns = [np.sum(self.returns * w) for w in portfolios]
      return portfolio_returns
  #this function calculates portfolios vols
  def portfolio_volatilities(self,portfolios):
      annualizing_factor = yearly_nb_periods()
      portfolio_volatilities = [np.sqrt(np.dot(w.T, np.dot(self.cov_matrix, w)))*np.sqrt(annualizing_factor) for w in portfolios]
      return portfolio_volatilities
````

## Capital Market Line (CML)



It is a line that represents a theoretical relationship between risk and return for efficient portfolios in the capital market.


### CML components and characteristics


Let's break down components and characteristics of the Capital Market Line:


>**Efficient Portfolios:**
The CML assumes that investors are rational and seek to maximize returns for a given level of risk.

>**Risk-Free Asset:**
The CML incorporates a risk-free asset. the asset that has no volatility or standard deviation of returns, making it a safe investment.

>**Efficient Frontier and Tangent Point:**
The CML is tangent to the efficient frontier where The point of tangency between is the optimal risky portfolio.

>**Equation:**
$$
E(R)=R_f+\frac{E\left(R_M\right)-R_f}{\sigma_M} \cdot \sigma
$$
where:
- $E(R)$ is the expected return of the portfolio,
- $R_f$ is the risk-free rate,
- $E\left(R_M\right)$ is the expected return of the market portfolio,
- $\sigma_M$ is the standard deviation (volatility) of the market portfolio,
- $\sigma$ is the standard deviation of the portfolio.

### Basic concepts

Durring this guide we will construct three different portfolios:

## Diversification


### Basic concepts


Diversification is achieved when the correlation between assets is not perfectly positive, indicating that the assets do not move in lockstep. Two key metrics, mean covariance and mean variance, provide insights into the quality and risk contributions of a diversified portfolio.
>**Mean covariance:** It offers valuable information about the quality of diversification within a portfolio. It is calculated as the average covariance between pairs of assets, considering all possible combinations. The formula is given by:
\begin{equation}
\overline{\operatorname{cov}}=\frac{1}{n(n-1)}\left(\sum_{i=1}^n \sum_{j=1}^n \operatorname{cov}\left(\tilde{r}_i, \widetilde{r}_j\right)\right)
\end{equation}
* Diversification benefits are enhanced when the mean covariance is minimized, signifying lower interdependence among asset returns.

>**Mean variance:** It is a measure of the risk contributed by multiple assets within a portfolio. It calculates the average variance of individual asset returns, providing insight into the overall risk profile. The formula is given by:
\begin{equation}
\overline{\sigma^2}=\frac{1}{n}\left(\sum_{i=1}^n \sigma_i^2\right)
\end{equation}
- A lower mean variance indicates that the risk associated with the portfolio is spread across multiple assets, emphasizing the potential benefits of diversification.


### Mean covariance
>Interdependence among asset returns

#### Calculation


````python
# Calculate mean covariance
def diversification_mean_covariance(close_df):
  mean_covariance = close_df.cov().mean().mean()
  return mean_covariance
````

## The model








This is a financial asset valuation model, namely the CAPM (Capital Asset Pricing Model). It was developed by Sharpe (1964), Lintner (1965), and Mossin (1966).
The CAPM is derived from the portfolio management results initiated by Markowitz in 1952. It relies on the concept of the efficient frontier and is based on assumptions related to portfolio management, supplemented by assumptions related to market operations.



###Model concepts

**The agents:**
Market participants must:
* Invest over a specific and common period
* Maximize their utility
* Be risk-averse.
* Have homogeneous expectations regarding asset characteristics: the variance-covariance matrix and the return vector are common to all participants.

**The market**: The market must be perfect in the sense that:
Information is free and accessible to all.
There are no transaction costs, whether buying or selling. This is contrary to reality but does not alter the general spirit of the model. There are no taxes. Securities are perfectly divisible and can be bought or sold short.

**The market portfolio**: The market portfolio is defined as the portfolio composed of all assets present in the market. Each asset is taken in proportion to its representation within the market (We only use BTC and ETH).


**Two-Fund Separation Theorem**:
This theorem involves separating funds into two categories, one without risk and the other risky:
* All investors optimally hold a portfolio composed of the risk-free asset and the market portfolio M.
\begin{equation}
\mathrm{P}^*=\lambda \mathrm{M}+(1-\lambda) r_f
\end{equation}
* The market portfolio, containing all the risky assets in the market, will be fully diversified. It will only capture the systematic risk of the market's risky assets.

### Linear Regression Model

**Linear Regression Model**: We perform linear regression of the asset's return relative to the market's return based on N observations.
\begin{equation}
r_i^t=a_i+b_i r_M^t+\varepsilon_i^t \quad \forall t=1, \ldots, N
\end{equation}


* **Regression Coefficients:**
The Ordinary Least Squares (OLS) method is used to obtain the coefficients of the regression. They are given by:
$$
\left\{\begin{array}{l}
b_i=\frac{\overline{\operatorname{Cov}\left(r_i, r_M\right)}}{\overline{\sigma_M^2}} \\
a_i=\overline{r_i}-b_i \overline{r_M}
\end{array}\right.
$$
Where :
  - $\bar{r}_i=\frac{1}{N} \sum_{t=1}^N r_i^t \quad$ : The empirical return of the asset,
  - $\overline{r_M}=\frac{1}{N} \sum_{t=1}^N r_M^t \quad:$ The empirical return of the market
  - $\overline{\operatorname{Cov}\left(r_i, r_M\right)}=\frac{1}{N-1} \sum_{i=1}^N\left(r_i^{\prime}-\bar{r}_i\right)\left(r_M^{\prime}-\bar{r}_{\mathrm{M}}\right):$ empirical market asset covariance  
  - $\overline{\sigma_M^2}=\frac{1}{N-1} \sum_{t=1}^N\left(r_M^t-\overline{r_M}\right)^2 \quad:$  empirical marketvariance


**Beta estimation**: The coefficient obtained in the regression of the returns of the asset against those of the market corresponds to the estimation of $\frac{\overline{\operatorname{Cov}\left(r_i, r_M\right)}}{\overline{\sigma_M^2}}$. Which is the estimation of the beta of the considered asset : \begin{equation}
\operatorname{Cov}\left(\widetilde{r}_i, \widetilde{r}_M\right)=\operatorname{Cov}\left(a_i+b_i \widetilde{r}_M+\widetilde{\varepsilon}_i, \widetilde{r}_M\right) \\
\Leftrightarrow \operatorname{Cov}\left(\widetilde{r}_i, \widetilde{r}_M\right)=b_i \operatorname{Cov}\left(\widetilde{r}_M, \widetilde{r}_M\right)+\operatorname{Cov}\left(\widetilde{\varepsilon}_i, \widetilde{r}_M\right) \\
\Leftrightarrow \operatorname{Cov}\left(\widetilde{r}_i, \widetilde{r}_M\right)=b_i \operatorname{Cov}\left(\widetilde{r}_M, \widetilde{r}_M\right) \operatorname{car} \operatorname{Cov}\left(\widetilde{\varepsilon}_i, \widetilde{r}_M\right)=0 .\end{equation}
\begin{equation}
\text { Finally } b_i=\frac{\operatorname{Cov}\left(\widetilde{r}_i, \widetilde{r}_M\right)}{\operatorname{Cov}\left(\widetilde{r}_M, \widetilde{r}_M\right)}=\frac{\operatorname{Cov}\left(\widetilde{r}_i, \widetilde{r}_M\right)}{\operatorname{Var}\left(\widetilde{r}_M\right)}
\end{equation}
This means that the slope of the linear regression will be an estimation of the beta.

## Beta and SML (Security Market Line)




### Beta basic concepts
>It is a measure of sensitivity which reflects the influence of the market on the asset's price and, consequently, represents systematic risk.

**Beta concept**:
The beta of an asset is defined as the ratio of its covariance with the market to the variance of the market.
This is the contribution of the asset to the total risk of the market. It establishes the relationship between the market and the considered asset.

**Beta formula**:
\begin{equation}
\beta_{\text {Asset }}=\frac{\operatorname{Cov}\left(\widetilde{r}_{\text {Market }}, \widetilde{r}_{\text {Asset }}\right)}{\operatorname{Var}\left(\widetilde{r}_{\text {Market }}\right)}
\end{equation}

**Assets classification**:  The beta coefficient highlights three classes of assets:
* **Aggressive** assets with a beta greater than 1
* **Defensive** assets with a beta less than 1
* **Neutral** assets with a unitary beta (assets behaving like the market)

#### Regression model implementation
>All risks calculation

**Implementation**

````python
def asset_risk(market_index,close_df,asset):
    df = pd.DataFrame(index=close_df.index)
    df= close_df.pct_change().bfill()
    index_return = market_index.pct_change().bfill()
    df['market_index']=index_return['market_index']
    # Simple Regression model
    # r = a + rb + Îµ
    formula = asset + '~market_index'
    lm = smf.ols(formula=formula, data=df).fit()
    # Residual variance
    Resid_model_var = lm.resid.var()
    # Market Variance
    var_m = df['market_index'].var()
    # Asset beta
    beta_asset = lm.params[1]
    # systematic risk
    systematik_risk = np.sqrt(var_m * beta_asset ** 2)
    # specific risk
    specific_risk = np.sqrt(Resid_model_var)
    # risque totale
    total_risk = np.sqrt(specific_risk ** 2 + systematik_risk ** 2)
    return {'Asset' : asset,
            'Beta_asset': beta_asset,
            'Class':'Agressive' if beta_asset>1 else 'Defensive',
            'Total_risk': round(total_risk,4) * 100,
            'Systematik_risk': round(systematik_risk,4) * 100,
            'Specific_risk': round(specific_risk,4) * 100
            }
````

### Security market line (SML)
>It is a graphical representation of the CAPM.

#### SML basic concepts

**SML concept**: It depicts the relationship between the expected return and the systematic risk (beta) of an investment in the context of the overall market  
It helps investors assess whether a given security or portfolio offers an appropriate expected return based on its risk, particularly its systematic or market risk.


 **SML Equation** :
$$
E(R)=R_f+\beta \times\left(E\left(R_m\right)-R_f\right)
$$

Where:
- $E(R)$ is the expected return on the investment.
- $R_f$ is the risk-free rate.
- $\beta$ is the beta of the investment.
- $E\left(R_m\right)$ is the expected market return.

 **Asset required return**: At market equilibrium,  the required return for the asset is equal to that of the risk-free asset plus a risk premium.
$$
\Pi_{\mathrm{Ai}}=\beta_i\Pi_{\mathrm{M}}
$$
Where
* $r_f$ : Risk free rate
* $\Pi_{\mathrm{M}}=E\left(\widetilde{r}_M\right)-r_f$ : Market risk primum
* $\Pi_{\mathrm{Ai}}=E\left(\widetilde{r}_i\right)-r_f$ : Asset Ai risk primum




>The SML is applicable to all assets, whereas the CML was only relevant to efficient portfolios.

#### SML simulation

````python
# Function to calculate SML
def sml(risk_free_rate, market_return, beta):
    return risk_free_rate + beta * (market_return - risk_free_rate)
````

### Treynor ratio

**Proprieties:**
- One dimension: profitability
- Two components: market fluctuations, portfolio fluctuations.
- Risk is measured by the beta coefficient.
- Treynor's T Index represents the risk premium per unit of risk (systematic).

**Formula:**
\begin{equation}
T=\frac{E\left(\widetilde{r}_P\right)-r_f}{\beta_P}
\end{equation}

**Implementation**

````python
#Treynor ratio
def treynor_ratio(portfolio,asset_risks,rf=0):
  portfolio_return = portfolio['Returns']
  portfolio_beta = portfolio_beta_cal(portfolio,asset_risks)
  t=(portfolio_return-rf)/portfolio_beta
  return t
````

### Sharpe ratio


**Proprieties:**
- Two dimensions: Profitability and diversification
- Risk is measured by the standard deviation.
- Sharpe's S Index represents the risk premium (total) per unit of risk.

**Formula:**
\begin{equation}
S=\frac{E\left(\widetilde{r}_P\right)-r_f}{\sigma_P}
\end{equation}

**Implementation**

````python
#Shape ratio
def shape_ratio(portfolio,rf=0):
  portfolio_return = portfolio['Returns']
  portfolio_std= portfolio['Volatility']
  s=(portfolio_return-rf)/portfolio_std
  return s
````


### Jensen index


**Proprieties:**
* Jensen's J Index represents the difference between
expected return and the required return by CAPM.

**Formula:**
\begin{equation}
J=E\left(\widetilde{r}_P\right)-\left(r_f+\beta\left[E\left(\widetilde{r}_M\right)-r_f\right]\right)
\end{equation}

**Implementation**

````python
#Jensen Indice
def jensen_indice(portfolio,market_return,asset_risks,rf=0):
  portfolio_beta = portfolio_beta_cal(portfolio,asset_risks)
  portfolio_return = portfolio['Returns']
  j = portfolio_return - (rf + portfolio_beta*(market_return-rf))
  return j
````
